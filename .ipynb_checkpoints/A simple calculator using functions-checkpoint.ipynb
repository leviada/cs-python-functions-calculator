```python
#makes the Menu and all of its valid inputs
def mainMenu():
    validChoices = ["A", "S","T","P","M"]
    valid = False
    while not valid:
        print("A)dd numbers")
        print("S)ubtract numbers")
        print("T)imes numbers")
        print("P)ower number")
        print("M)od number")
        choice = input("choose your option").upper()
        if choice not in validChoices:
            print("sorry that is not a valid menu choice ")
        else:
            valid= True
            
    return choice
    
#checks the number that is inputed
def enterNumber():
    valid = False
    while not valid:
        try:
            x=int(input("enter the first number"))
            valid=True
        except ValueError:
            print("That was not a valid number.. please try again")
    
    return x
#functions for adding
def div(x,y):
    
    return x/y

def mod(x,y):
    
    return x%2

def power(x,y):
    
    return x**y

def times(x,y):
    
    return x * y

def add(x,y):
    
    return x+y

def subtract(x,y):
    
    return x-y

#main program starts here
print("welcome to my calculator")
again = "y"
while again == "y":
    
    userChoice = mainMenu()
    if userChoice == "M":
        answer = mod(enterNumber(), enterNumber())
    elif userChoice == "T":
        answer = times(enterNumber(), enterNumber())
    elif userChoice == "A":
        answer = add(enterNumber(), enterNumber())
    elif userChoice == "S":
        answer = subtract(enterNumber(), enterNumber())
    elif userChoice == "P":
        answer = power(enterNumber(), enterNumber())
    else:
        print("error - we should never get here")
    
    print("the answer from your calcaulation is="+str(answer))
    
```
